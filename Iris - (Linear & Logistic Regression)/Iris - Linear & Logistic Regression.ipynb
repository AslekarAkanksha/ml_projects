## - Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("target",axis=1)
y = df['target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#  Linear Regression :-

```python
from sklearn.linear_model import LinearRegression
```

```python
LR_model = LinearRegression()
```

```python
LR_model.fit(x_train,y_train)
```

```python
LR_model.coef_                                       ## m ---> slope    [for each feature[/column]]
```

```python
LR_model.intercept_                                  ## b .... intercept
```

### Training data evaluation

- **Actual** ------ <u>y - train</u>

```python
y_train.head()
```

- **Predicted** ------ <u>y - train</u>

```python
y_train_pred = LR_model.predict(x_train)
y_train_pred

## above o/p is in array so convert it into series to apply head()
pd.Series(y_train_pred).head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(y_train,y_train_pred)
mae = mean_absolute_error(y_train,y_train_pred)
R2 = r2_score(y_train,y_train_pred)

print("MSE     :",mse)
print("MAE     :",mae)
print("R2-score  :",R2)
```

### Testing data evaluation

- **Actual** ------ <u>y - test</u>

```python
y_test.head()
```

- **Predicted** ------ <u>y - test</u>

```python
y_test_pred = LR_model.predict(x_test)
y_test_pred

## above o/p is in array so convert it into series to apply head()
pd.Series(y_test_pred).head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(y_test,y_test_pred)
mae = mean_absolute_error(y_test,y_test_pred)
R2 = r2_score(y_test,y_test_pred)

print("MSE     :",mse)
print("MAE     :",mae)
print("R2-score  :",R2)
```

# Logistic Regression (Multiclass)

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target'],axis=1)
x.head(2)
```

```python
y = df['target']
y.head(2)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model = LogisticRegression(multi_class='ovr')
```

```python
lg_model.fit(x_train,y_train)
```

## Training Data Evaluation :

```python
y_train.head(2)
```

```python
y_train_p = lg_model.predict(x_train)
y_train_p
```

```python
y_train_pred = pd.Series(y_train_p)
y_train_pred.head(2)
```

```python
acc = accuracy_score(y_train,y_train_pred)
acc
```

```python
clf_report = classification_report(y_train,y_train_pred)
print(clf_report)
```
